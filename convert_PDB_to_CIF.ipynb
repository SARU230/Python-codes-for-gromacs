```python
import os
from rdkit import Chem
from ase import io

def convert_pdb_to_cif(pdb_file):

    try:
        structure = io.read(pdb_file)
        cif_file = os.path.splitext(pdb_file)[0] + '.cif'
        structure.write(cif_file)
        print(f"Converted {pdb_file} to {cif_file}")
    except Exception as e:
        print(f"Error converting {pdb_file}: {e}")

def process_files_in_directory():
    directory = os.getcwd()
    for filename in os.listdir(directory):
        if filename.startswith("Tips_") and filename.endswith(".pdb"):
            pdb_path = os.path.join(directory, filename)
            convert_pdb_to_cif(pdb_path)
#destination="F:/OneDrive - University of Kentucky/UK work/Research/Karl/TIPS_PN/Tips_Pen_293K_PDB"
process_files_in_directory()
```
